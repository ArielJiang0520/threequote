```python
from Scweet.scweet import scrap
from Scweet.user import get_user_information, get_users_following, get_users_followers

data = scrap(
    words=['snapchat%20(bug%20OR%20error%20OR%20issue)'], 
    start_date="2021-01-01", 
    max_date="2021-03-04",
    from_account = None,
    interval=1,
    headless=True, 
    display_type="Latest", 
    save_images=False,
    resume=False, 
    limit=10,
    filter_replies=False, 
    proximity=False)
```

```python
data.head()
```

```python
import pickle

pickle.dump(data, open('snapchat.p', 'wb'))
```
